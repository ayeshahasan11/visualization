```python

# generate random sample data
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

# insert a histogram
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y,bins=10)

# plot labels and titles
ax.set_title('Total growth over time')
ax.set_xlabel('Total growth')
ax.set_ylabel('Years since start')
```

```python
#dictionary with properties we want for our labels
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

#plot labels and titles
ax.set_title('Total growth over time', fontdict = font1)
ax.set_xlabel('Total growth', fontdict = font2)
ax.set_ylabel('Years since start', fontdict = font2)
```

```python
# insert scatter plot with different marker and color
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y, marker = '*', color = 'indigo')
```

```python
# adjust style and add connecting lines to the scatter plot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y, 
            marker = '*', 
            markersize = 12,
            color = 'indigo',
            linestyle = '--',
            linewidth = 2,
            markeredgecolor = '#fa9359',
            markerfacecolor = '#000000')

# add gridlines to the plot
ax.grid(axis = 'y', 
color = 'gray',
        linestyle = '--',
        linewidth = 0.5,)
```

```python
# Activity

# libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
 
# Dataset:
a = pd.DataFrame({ 'group' : np.repeat('A',500), 'value': np.random.normal(10, 5, 500) })
b = pd.DataFrame({ 'group' : np.repeat('B',500), 'value': np.random.normal(13, 1.2, 500) })
c = pd.DataFrame({ 'group' : np.repeat('B',500), 'value': np.random.normal(18, 1.2, 500) })
d = pd.DataFrame({ 'group' : np.repeat('C',20), 'value': np.random.normal(25, 4, 20) })
e = pd.DataFrame({ 'group' : np.repeat('D',100), 'value': np.random.uniform(12, size=100) })
df = pd.concat([a,b,c,d,e])

# plot violin chart
sns.violinplot( x='group', y='value', data=df, color='olive')
sns.set_style("whitegrid")

# add title
plt.title("Violin plot", loc="left")

# add gridlines
plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5)

# show the graph
plt.show()
```
